```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
td = [6,12.5,19,25,37,42.5,48.5,53,57,61.5,66,71,76.5]
tr=[6,12.5,17,19.5,25,30.5,37.5,45.5,52,59,65.5,71,76.5]
receptor = np.asarray([12.5,25,37.5,50,75,100,125,150,175,200,225,250,275])
```

```python
receptor2 = 275 - rec
```

```python
plt.figure(figsize=(12,10))
plt.plot(receptor,td,marker='o',lw=2,label='Tempo de trnsito - Forward')
plt.plot(receptor2,tr,marker='X',lw=2,ls='--',label='Tempo de trnsito - Reverso')
plt.ylabel('Tempo ($m s$)',fontsize=16)
plt.xlabel('Receptor (m)',fontsize=16)
plt.title('Sismograma sinttico',fontsize=18)
plt.xticks(np.arange(0,300,25))
plt.grid()
plt.gca().invert_yaxis()
plt.legend()
plt.savefig('grfico-questo-4-cor.png',dpi=300)
```

### Anlise do levantamento direto

```python
coefd = []
for i in range(0,len(td)-1):
    coefd.append((td[i+1]-td[i])/(receptor[i+1]-receptor[i]))
```

```python
print(coefd)
```

```python
def reta(x,x0,y0,alpha):
    return alpha*x - alpha*x0 + y0

def direta(x,alpha):
    return x*alpha
```

```python
plt.figure(figsize=(12,10))
plt.plot(receptor,td,marker='o',lw=2,label='Tempo de trnsito - Forward')

plt.plot(receptor[0:5],direta(receptor[0:5],0.52),ls='--',label='direta')

plt.plot(receptor[2:6],reta(receptor[2:6],receptor[2],td[2],0.48),label='refratada 1')

plt.plot(receptor[4:9],reta(receptor[4:9],receptor[4],td[4],0.22),label='refratada 2')

plt.plot(receptor[7:],reta(receptor[7:],receptor[8],td[8],coefd[8]),label='refratada 3')

plt.plot(receptor[9:],reta(receptor[9:],receptor[-1],td[-1],coefd[-1]),label='refratada 4',color='orange')

plt.ylabel('Tempo ($m s$)',fontsize=16)
plt.xlabel('Receptor (m)',fontsize=16)
plt.title('Sismograma sinttico',fontsize=18)
plt.xticks(np.arange(0,300,25))
plt.grid()
plt.gca().invert_yaxis()
plt.legend()
#plt.savefig('grfico-questo-4-cor.png',dpi=300)
```
